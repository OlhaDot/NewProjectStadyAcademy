```python
# task1
import pandas as pd
import numpy as np
import networkx as nx
```

```python
see = pd.read_csv(r"C:\Users\Asus\PycharmProjects\NewProjectStadyAcademy\cities.csv")
```

```python
cities = np.array(see)
```

```python
G = nx.Graph()
for edge in cities:
    G.add_edge(edge[0],edge[1], weight = edge[2])
nx.draw(G)
```

```python
# task2
def the_shortest_way(graph, city1, city2):
    print("Itinerary: ", nx.shortest_path(graph, city1, city2, weight='weight'))
    print("Duration of itinerary:", nx.shortest_path_length(graph, city1, city2, weight='weight'))

the_shortest_way(G, 'Odesa', 'Zinkiv')
```
