```python
# find somm by course, range, by city
# aggregation
df.groupby(['title', 'city'], as_index = False)\
  .agg({'money': 'sum', 'number': 'count'})\
  .sort_values('money', ascending= False)
```

```python
# memorize
money_by_city = df.groupby(['title', 'city'], as_index = False)\
  .agg({'money': 'sum', 'number': 'count'})\
  .sort_values('money', ascending= False)
```

```python
# write file to particular direction...?
money_by_city.to_csv('https://drive.google.com/drive/folders/14CUhcz6DZC0TFTg7isie3FwWw4f1m81R?usp=drive_link/money_to_csv',  index = False)
```

```python
# f-tion where...
cnt = df\
  .query('number >0')\
  .groupby(['title', 'city'], as_index = False)\
  .agg({'money': 'sum', 'number': 'count'})\
  .sort_values('money', ascending= False)
cnt
```

```python
# where
finished_cources = df\
  .query('status == "Завершен"')\
  .groupby(['title', 'city'], as_index = False)\
  .agg({'money': 'sum', 'number': 'count'})\
  .sort_values('number', ascending= False)
finished_cources
```

```python
# renume columns
finished_cources_succ = df\
  .query('status == "Завершен"')\
  .groupby(['title'], as_index = False)\
  .agg({'money': 'sum', 'number': 'count'})\
  .sort_values('number', ascending= False)\
  .rename(columns= {'number': "success_cources"})
finished_cources_succ
```

```python
# check
if df.money.sum() == finished_cources_succ.money.sum():
  print("ok")
else:
  print("error!!")
```
