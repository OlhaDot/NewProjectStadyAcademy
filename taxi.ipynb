```python
# Сгруппируйте данные по двум признакам:
#  району города и является ли день выходным (колонки borough и hday). Сравните среднее число поездок, и выберите районы,
#  из которых по праздникам в среднем поступает больше заказов,
# чем в обычные дни.
```

```python
taxi\
.groupby(["borough", "hday"], as_index=False)\
.agg({"pickups": "mean"})\
.sort_values("pickups", ascending=False)
```

```python
"""Для каждого района посчитайте число поездок по месяцам. Отсортируйте полученные значения по убыванию и
сохраните результирующий датафрейм в pickups_by_mon_bor."""
```

```python
pickups_by_mon_bor = taxi.groupby(["pickup_month", "borough"], as_index=False)\
.agg({"pickups": "sum"}).sort_values("pickups", ascending=False)
```

```python
def temp_to_celcius(temp_f):
    temp_c = (temp_f - 32) * 5.0 / 9.0
    return pd.Series(temp_c, name='temp_C')
```

```python
 taxi['temp_c'] = temp_to_celcius(taxi['temp'])
```

```python
def temp_to_celcius2(temp_f):
    temp_c = (temp_f - 32) * 5.0 / 9.0
    return temp_c
```

```python
 taxi['temp_c2'] = temp_to_celcius2(taxi['temp'])
```
