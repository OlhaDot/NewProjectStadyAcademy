```python
import pandas as pd
import numpy as np

# %matplotlib ipympl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# task 1 read csv file 
file = pd.read_csv(r'/content/drive/MyDrive/Pyton_DS/lesson33/cars.csv')
print(file)
```

```python
# task 2 - line graph
plt.plot(file['price'], file['consumption'])
plt.xlabel('Car price')
plt.ylabel('Fuel consumption')
plt.title('Line chart of fuel consumption vs price')
plt.show()
```

```python
# task 2.2 - babble graph
plt.scatter(file['speed'], file['consumption'])

plt.xlabel('Speed (km/h)')
plt.ylabel('Fuel consumption')
plt.title('Scatter chart of fuel consumption vs speed')

plt.show()

```

```python
# task 2.3 - babble graph
plt.scatter(file['capacity'], file['speed'])

plt.xlabel('Car capacity')
plt.ylabel('Speed (km/h)')
plt.title('Scatter chart of car capacity vs speed')

plt.show()
```

```python
# task 3 - histogram with 5 bins
plt.hist(file['speed'], bins=5)

plt.title("Distribution of Car Speeds")
plt.xlabel("Speed (km/h)")
plt.ylabel("Frequency")

plt.show()
```

```python
# task4 -boxplot
# assumptions about car brand speeds
audi_speeds = [120, 130, 145, 155, 165, 170, 180, 185, 210, 220, 240, 250]
bmw_speeds = [115, 125, 135, 150, 155, 165, 175, 180, 190, 200, 210, 220]
mercedes_speeds = [110, 120, 130, 140, 150, 155, 165, 170, 175, 190, 215, 230]

fig, ax = plt.subplots()
ax.boxplot([audi_speeds, bmw_speeds, mercedes_speeds])

ax.set_title('Speeds distribution by car brand')
ax.set_xticklabels(['Audi', 'BMW', 'Mercedes'])
ax.set_ylabel('Speed (km/h)')

plt.show()
```
